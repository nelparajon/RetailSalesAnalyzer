```python
#comprobar si hay valores nulos y cuantos
nulls = retail_df.isnull().sum()
print(nulls)
```

```python
#comprobar si hay duplicados
duplicated = retail_df.duplicated().sum()
print(duplicated)
```

```python
#eliminamos los valores nulos
retail_df.dropna()
```

```python
#eliminamos los valores duplicados
retail_df.drop_duplicates()
```

```python
#ordenamos por categoria de producto
products_by_category = retail_df.sort_values(by="Product Category")
print(products_by_category)
```

```python
#ordenamos por fecha
sort_by_date = retail_df.sort_values(by="Date", ascending=True)
print(sort_by_date)
```

```python
#filtramos el total de ganancias por genero y mostramos solamente gender y amount y un porcentaje
amount_by_gender = retail_df.groupby("Gender")["Total Amount"].sum().reset_index()
total_amount = retail_df["Total Amount"].sum()
amount_by_gender["Porcentaje"] = (amount_by_gender["Total Amount"] / total_amount) * 100
print(amount_by_gender[["Gender", "Total Amount", "Porcentaje"]])
```

```python
#filtramos por mes y genero para saber que meses compran mas mujeres y que mes compran mas hombres en porcentaje
#primero convertimos el tipo de "Date" a un tipo válido, ya que ahora mismo está en object
retail_df["Date"] = pd.to_datetime(retail_df["Date"], errors='coerce')
#creamos una nueva columna que sea año-mes YYYY-MM
retail_df["Year-Month"] = retail_df["Date"].dt.to_period("M")
#calculamos el total amount para cada genero por mes
amount_by_gender_month = retail_df.groupby(["Year-Month", "Gender"])["Total Amount"].sum().reset_index()
#calculamos el total amount mensual
total_by_month = retail_df.groupby("Year-Month")["Total Amount"].sum().reset_index()
total_by_month = total_by_month.rename(columns={"Total Amount": "Total Monthly Amount"})
#combinamos ambos df
merge_df = pd.merge(amount_by_gender_month, total_by_month, on="Year-Month")
#calculamos el porcentaje de comprar al mes por genero y lo mostramos en columna
merge_df["Porcentaje"] = (merge_df["Total Amount"] / merge_df["Total Monthly Amount"]) * 100
print(merge_df)
```

```python
#filtramos el total por category para conocer en que se gasta mas al año
amount_by_category = retail_df.groupby("Product Category")["Total Amount"].sum()
print(amount_by_category)
```

```python
import matplotlib.pyplot as plt

# Filtrar el total por categoría de producto para conocer en qué se gasta más
amount_by_category = retail_df.groupby("Product Category")["Total Amount"].sum()

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))  # Establecer el tamaño del gráfico

# Generar gráfico de pastel
plt.pie(amount_by_category, labels=amount_by_category.index, autopct='%1.1f%%', startangle=90)

# Añadir título al gráfico
plt.title('Distribuciones Anuales por Categoría de Producto')

# Asegurarse de que el gráfico sea un círculo
plt.axis('equal')  

# Mostrar el gráfico
plt.show()
```

```python
#filtramos el total por dias
amount_by_date = retail_df.groupby("Date")["Total Amount"].sum()
print(amount_by_date)
```

```python
#filtramos el total por meses usando el método Grouper. Date como key y el mes (M) como la frecuencia a filtrar
#convertimos el tipo de dato de Date a datetime
retail_df["Date"] =   pd.to_datetime(retail_df["Date"], errors='coerce')

#calculamos y filtramos el total amount por mes
amount_by_month = retail_df.groupby(pd.Grouper(key="Date", freq="ME"))["Total Amount"].sum().reset_index()
print(amount_by_month)

plt.figure(figsize=(10,6))
plt.plot(amount_by_month["Date"], amount_by_month["Total Amount"], marker='o', linestyle='-', color='b')
plt.title('Subidas y Bajadas de Total Amount por Mes')
plt.xlabel('Fecha (Mes)')
plt.ylabel('Total Amount')

# Rotar las etiquetas del eje X para mayor legibilidad
plt.xticks(amount_by_month["Date"], rotation=45)

# Mostrar el gráfico
plt.show()
```

```python
# Asegúrate de que la columna "Date" esté en formato datetime y crear la columna "Month"
retail_df["Date"] = pd.to_datetime(retail_df["Date"], errors='coerce')
retail_df["Month"] = retail_df["Date"].dt.to_period("M")

# Agrupar y calcular el total por cada categoría y el total mensual
amount_by_cat_month = retail_df.groupby(["Month", "Product Category"])["Total Amount"].sum().reset_index()

# Calcular el total mensual
amount_by_month = retail_df.groupby("Month")["Total Amount"].sum().reset_index()

# Combinar ambos DataFrames
merge_df = pd.merge(amount_by_cat_month, amount_by_month, on="Month", suffixes=('_Cat', '_Total'))

# Calcular el porcentaje por categoría y mes
merge_df["Porcentaje"] = (merge_df["Total Amount_Cat"] / merge_df["Total Amount_Total"]) * 100

# Filtrar las filas con el porcentaje máximo por cada mes
max_porcentaje_idx = merge_df.groupby("Month")["Porcentaje"].idxmax()
categorias_max_vendidas = merge_df.loc[max_porcentaje_idx]

# Preparar los datos para el gráfico
resultado = categorias_max_vendidas[["Month", "Product Category", "Porcentaje"]]

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))  # Establecer el tamaño del gráfico
bars = plt.bar(resultado["Month"].astype(str), resultado["Porcentaje"], color='skyblue')

# Añadir título y etiquetas de los ejes
plt.title('Categoría de Producto con Mayor Porcentaje de Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Porcentaje de Ventas')

# Mostrar las etiquetas de categoría y porcentaje sobre las barras
for index, bar in enumerate(bars):
    yval = bar.get_height()
    category = resultado["Product Category"].iloc[index]
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{category}\n{yval:.2f}%", 
             ha='center', va='bottom', fontsize=10)

# Rotar las etiquetas del eje X para mejor legibilidad
plt.xticks(rotation=45)

# Ajustar el gráfico para que no se corten las etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
